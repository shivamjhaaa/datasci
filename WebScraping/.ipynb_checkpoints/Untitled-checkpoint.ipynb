## Web Scraping

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://leetcode.com/problemset/all/?page='
ans = []
for i in range(1,5):
    print(url + f'{i}')
    response = requests.get(url + f'{i}').text
    soup = BeautifulSoup(response)

    questions = soup.find_all('a',class_='h-5')
    print(questions,'\n\n')

```
